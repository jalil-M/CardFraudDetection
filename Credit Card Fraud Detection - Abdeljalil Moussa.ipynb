# Introduction

The datasets contains transactions made by credit cards in September 2013 by european cardholders.This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions.

It contains only numerical input variables which are the result of a PCA transformation.

Due to confidentiality reasons, we do not get further information on the features, presented below :

- Features V1, V2, ... V28 are the principal components obtained with PCA;
- The only features which have not been transformed with PCA are Time and Amount. Feature Time contains the seconds elapsed between each transaction and the first transaction in the dataset. The feature Amount is the transaction Amount, this feature can be used for example-dependant cost-senstive learning.
- Feature Class is the response variable and it takes value 1 in case of fraud and 0 otherwise.

The dataset has been collected and analysed during a research collaboration of Worldline and the Machine Learning Group (http://mlg.ulb.ac.be) of ULB (Universit√© Libre de Bruxelles) on big data mining and fraud detection.

## Pre-processing Methods

This part represents the pre-processing analysis on our dataset :

- we will first scale the columns comprise of Time and Amount . Time and amount should be scaled as the other columns. 
- Furthermore, we need to also create a sub sample of the dataframe in order to have an equal amount of Fraud and Normal cases, helping our algorithms better understand patterns that determines whether a transaction is a fraud or not.

```python
from sklearn.preprocessing import StandardScaler, RobustScaler

# Since we have some outliers in our dataset, it is better to use the Robust Scaler method from scikit-learn.

robust = RobustScaler()

df['norm_amount'] = robust.fit_transform(df['Amount'].values.reshape(-1,1))
df['norm_time'] = robust.fit_transform(df['Time'].values.reshape(-1,1))

df.drop(['Time','Amount'], axis=1, inplace=True)
```

_In this step of the pre-processing part, we will implement **"Random Under Sampling"** which consists of removing data randomly in order to have a perfectly balanced dataset in order to the **overfitting** problem.
The overfitting problem is the main problem with imbalanced data because it will assume that in most cases there are no frauds. We do not want that!_

```python
df = df.sample(frac=1)

# From the class.describe() function above, we obtain that the number of fraud transactions is 492
fraud_df = df.loc[df['Class'] == 1]
normal_df = df.loc[df['Class'] == 0][:492]

std_df = pd.concat([fraud_df, normal_df])

# Shuffle dataframe rows in order to see if our model maintains a certain accuracy
df2 = std_df.sample(frac=1, random_state=42)

df2.head()
```

```python
sns.countplot('Class', data=df2)
```

_This is satisfying result, since now the dataset is perfectly **balanced** (50-50). Our subsample is a dataframe with a 50/50 ratio of fraud and normal transactions. Therefore, our subsample will have the same amount of fraud and normal transactions._

```python
plt.figure(figsize = (10,10))
plt.title('Vanilla Pearson Correlation Matrix on Credit Card Transactions')
corr = df.corr()
sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns,linewidths=.1,cmap="Blues")
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.title('UnderSampled Pearson Correlation Matrix on Credit Card Transactions')
corr = df2.corr()
sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns,linewidths=.1,cmap="Blues")
plt.show()
```

_Comparing the two correlation matrices (before and after under-sampling) demonstrates how we can highlight the pearson correlations between features now.<br>
Having a balanced dataset also **improves** the final classification accuracy regardless of the machine learning model.
In conclusion, this pre-processing part enabled us to both highlight the important features and solve the scaling and overfitting problems at the same time._


## MACHINE LEARNING CLASSIFICATION

In this part, we will apply two machine learning models for our classification challenge :
- Random Forest Classifier
- Neural Networks using Keras MLP

The first model is an ensemble method using bootsraping <br>
The second model is a deep learning model.

Hence, we can compare the classification analysis on both models.

```python
from sklearn.model_selection import train_test_split

y = 'Class'
x = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',\
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19',\
       'V20', 'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28',\
       'norm_amount', 'norm_time']

train_df, test_df = train_test_split(df2, test_size=0.2, shuffle=True )
train_df, valid_df = train_test_split(df2, test_size=0.2, shuffle=True )
```

## Ensemble Model : Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_jobs=5,
                             n_estimators=100,
                             verbose=False)

model.fit(train_df[x], train_df[y].values)
preds = model.predict(valid_df[x])

result = pd.DataFrame({'Feature': x, 'Feature importance': model.feature_importances_})
result = result.sort_values(by='Feature importance',ascending=False)
plt.figure(figsize = (6,6))
plt.title('Feature Importance Plot on Random Forest Classifier')
s = sns.barplot(x='Feature',y='Feature importance',data=result)
s.set_xticklabels(s.get_xticklabels(),rotation=90)
plt.show() 
```

```python
confusion = pd.crosstab(valid_df[y].values, preds, rownames=['Real Value'], colnames=['Predicted Value'])
fig, axr = plt.subplots(ncols=1, figsize=(10,10))
sns.heatmap(confusion, 
            xticklabels=['Not Fraud', 'Fraud'],
            yticklabels=['Not Fraud', 'Fraud'],
            annot=True,ax=axr,
            cmap="Blues")
plt.title('Confusion Matrix on Random Forest Regression Model', fontsize=14)
plt.show()
```

_The accuracy of our model is actually outstanding : **93%**<br>
The Confusion matrix shows us that our model is good with true-positives and true-negatives. <br>
However we still struggle with the False negatives but this can be explained by the fact that some fraudulent transactions use "unusual" fraud techniques that the model cannot predict correctly._

## Deep Learning Model : Neural Networks Keras MLP

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(train_df[x])
train_df2 = scaler.transform(train_df[x])
test_df2 = scaler.transform(test_df[x])
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout

def curves_plot(history,train,validation):
    plt.plot(history.history[train])
    plt.plot(history.history[validation])
    plt.title('Train History')
    plt.ylabel(train)
    plt.xlabel('Epoch')
    plt.legend(['train', 'validation'], loc='best')
    plt.show()

model = Sequential()
model.add(Dense(units=250, 
                input_dim=30, 
                kernel_initializer='uniform', 
                activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=250,  
                kernel_initializer='uniform', 
                activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1,
                kernel_initializer='uniform',
                activation='sigmoid'))

print(model.summary())

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop', metrics=['accuracy'])

train_history = model.fit(x=train_df2, y=train_df[y],
                          validation_split=0.8, epochs=100, 
                          batch_size=500, verbose=2)

curves_plot(train_history,'acc','val_acc')
curves_plot(train_history,'loss','val_loss')

scores = model.evaluate(test_df2, test_df[y])
print('\n')
print('accuracy=',scores[1])

preds = model.predict_classes(test_df2)
```

_The Deep Learning model is also doing well with a final accuracy of **91%** on 100 epochs.
However, we can still improve our algorithm by tuning the hyper-parameters of the dense layers and changing the optimizers method for example._
